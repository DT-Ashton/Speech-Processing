```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sinusoidal_wave(A, omega, phi, N):
    n = np.arange(N)
    x = A * np.cos(omega * n + phi)
    return n, x

def compute_magnitude_spectrum(x):
    N = len(x)
    X = np.fft.fft(x)
    magnitude_spectrum = np.abs(X) / N
    return magnitude_spectrum
```

```python
A = 2  # Amplitude
omega = 0.1 * np.pi  # Angular frequency
phi = 0.5 * np.pi  # Phase shift
N = 100  # Number of samples

n, x = generate_sinusoidal_wave(A, omega, phi, N)

# Plotting the sinusoidal wave
plt.figure(figsize=(10, 4))
plt.stem(n, x, label='x[n]')
plt.title('Discrete-Time Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.show()

magnitude_spectrum = compute_magnitude_spectrum(x)
omega = 2 * np.pi * np.arange(N) / N  # Radiant per sample

# Plotting the magnitude spectrum
plt.figure(figsize=(10, 4))
plt.stem(omega[:N//2], magnitude_spectrum[:N//2], basefmt=" ")
plt.title('Magnitude Spectrum of Discrete-Time Signal')
plt.xlabel('Angular Frequency  (rad/sample)')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
